## Bibliotecas

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

## Leitura

```python
df = pd.read_csv("Python_M10_support material.csv")
```

## Processamento

```python
# verificando tipagem dos dados
df.dtypes
```

```python
transformar_em_float = ['qtd_transacoes_12m','valor_transacoes_12m']

for col in transformar_em_float:
    # Convertendo para string
    df[col] = df[col].astype(str)
    # substituindo . da casa de milhar por nada
    df[col] = df[col].str.replace(".","")
    # substituindo a v√≠rgula da casa de milhar por ponto
    df[col] = df[col].str.replace(",",".")
    # mudando o tipo para float
    df[col] = df[col].astype("float")
```

```python
df.dtypes
```

```python
df.head()
```

### Segmenta√ß√£o de clientes

```python
# Cores para os setores do gr√°fico de pizza
cores = ['pink', 'blue']
```

```python
df_por_sexo = df.groupby('sexo').agg({'id':'nunique'}).reset_index()
```

```python
# Cria√ß√£o do gr√°fico de pizza
plt.figure(figsize=(8, 6))
plt.pie(df_por_sexo['id'], labels=df_por_sexo['sexo'], colors=cores, autopct='%1.1f%%', startangle=140)
plt.axis('equal')

# t√≠tulo
plt.title('Distribui√ß√£o de clientes por sexo')

# Mostra o gr√°fico
plt.show()
```

```python
idades = df[['idade']]
```

```python
# Criar o histograma
plt.figure(figsize=(8, 6))
plt.hist(idades, bins=10, color='skyblue', edgecolor='black', alpha=0.7)

# Adicionar r√≥tulos e t√≠tulo
plt.xlabel('Idade')
plt.ylabel('Frequ√™ncia')
plt.title('Distribui√ß√£o de Idades dos Clientes')

# Mostrar o histograma
plt.show()
```

```python
df_por_escolaridade = df.groupby('escolaridade').agg({'id':'nunique'}).reset_index()
```

```python
# Criar o gr√°fico de barras
plt.figure(figsize=(10, 6))
plt.bar(df_por_escolaridade['escolaridade'], df_por_escolaridade['id'], color='grey')

# Adicionar r√≥tulos e t√≠tulo
plt.xlabel('Escolaridade')
plt.ylabel('N√∫mero de Clientes')
plt.title('Distribui√ß√£o de Clientes por Escolaridade')

# Rotacionar os r√≥tulos no eixo x para melhor legibilidade
plt.xticks(rotation=45, ha='right')

# Mostrar o gr√°fico de barras
plt.tight_layout()
plt.show()
```

```python
df_por_salario = df.groupby('salario_anual').agg({'id':'nunique'}).reset_index()
```

```python
# Criar o gr√°fico de barras
plt.figure(figsize=(10, 6))
plt.bar(df_por_salario['salario_anual'], df_por_salario['id'], color='green')

# Adicionar r√≥tulos e t√≠tulo
plt.xlabel('Sal√°rio Anual')
plt.ylabel('N√∫mero de Clientes')
plt.title('Distribui√ß√£o de Clientes por Sal√°rio Anual')

# Rotacionar os r√≥tulos no eixo x para melhor legibilidade
plt.xticks(rotation=45, ha='right')

# Mostrar o gr√°fico de barras
plt.tight_layout()
plt.show()
```
